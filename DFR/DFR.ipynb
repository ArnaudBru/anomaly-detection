```
import os
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import auc
os.listdir('models')
```

```
predictions = {}
path = os.path.join(os.getcwd(), 'models')
for category in os.listdir(path):
    if category == "metal_nut":
        continue
    cat_path = os.path.join(path, category)
    model = os.listdir(cat_path)[0]
    preds_path = os.path.join(cat_path, model + '/eval/predictions.npz')
    predictions[category] = np.load(preds_path)
```

```
auc_scores = []
categories = predictions.keys()

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

for category in predictions:
    fpr = predictions[category]['det_fpr']
    tpr = predictions[category]['det_tpr']
    auc_scores.append(round(auc(fpr, tpr), 3))
    plt.plot(fpr, tpr)
    
plt.legend([cat + f' (auc = {auc})' for cat, auc in zip(categories, auc_scores)])
plt.title("Detection ROC Curves")
plt.savefig("detection_roc.png")
plt.show()
```
